```python
# for real con pan queso y todos los chiches automatizados para 22 archivos. 

import csv
import json
import os

# Directorio donde se encuentran los archivos JSON
directory = r'C:\Users\PC\Downloads\archivos har y json\PelucaMilei'

# Lista para almacenar los datos estructurados de todos los archivos
all_structured_data = []

# Iterar sobre los archivos JSON en el directorio
for filename in os.listdir(directory):
    if filename.endswith(".json"):
        file_path = os.path.join(directory, filename)

        # Abrir el archivo JSON y cargarlo como un objeto JSON
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)

        # Lista para almacenar los datos estructurados de un archivo
        structured_data = []

        # Iterar sobre cada elemento en itemList
        for item in data['itemList']:
            # Crear un diccionario para almacenar los datos estructurados de cada elemento
            item_data = {}
            
            # Verificar si 'stats' está presente en el elemento actual
            if 'stats' in item:
                # Obtener la sección 'stats' del elemento
                stats = item['stats']
                
                # Extraer los valores de 'collectCount', 'commentCount', 'diggCount', 'playCount' y 'shareCount' de 'stats'
                collect_count = stats.get('collectCount')
                comment_count = stats.get('commentCount')
                digg_count = stats.get('diggCount')
                play_count = stats.get('playCount')
                share_count = stats.get('shareCount')
                
                # Obtener la sección 'contents' del elemento
                contents = item.get('contents', [])
                
                # Extraer la descripción del primer contenido (si existen contenidos)
                first_content_desc = contents[0].get('desc') if contents else None
                
                # Obtener la sección 'music' del elemento
                music = item.get('music')
                
                # Extraer el valor de 'authorname' de 'music'
                author_name = music.get('authorName') if music else None
                
                # Extraer el valor de 'duration' de 'music'
                duration = music.get('duration') if music else None
                
                # Almacenar los valores en el diccionario item_data
                item_data['Collect Count'] = collect_count
                item_data['Comment Count'] = comment_count
                item_data['Digg Count'] = digg_count
                item_data['Play Count'] = play_count
                item_data['Share Count'] = share_count
                item_data['Author Name'] = author_name
                item_data['Duration'] = duration
                item_data['First Content Desc'] = first_content_desc
                
                # Agregar el diccionario item_data a la lista structured_data
                structured_data.append(item_data)
        
        # Agregar los datos estructurados de un archivo a la lista de todos los datos estructurados
        all_structured_data.extend(structured_data)

# Escribir los datos estructurados de todos los archivos en un solo archivo CSV
csv_file_path = r'C:\Users\PC\Downloads\Proyectos\structured_data.csv'
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Collect Count', 'Comment Count', 'Digg Count', 'Play Count', 'Share Count', 'Author Name', 'Duration', 'First Content Desc']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    for item_data in all_structured_data:
        writer.writerow(item_data)
```

```python
#alternativooo con estructuracion de data

import json

file_path = r'C:\Users\PC\Downloads\raw_response.json'

# Abrir el archivo JSON y cargarlo como un objeto JSON
with open(file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Lista para almacenar los datos estructurados
structured_data = []

# Iterar sobre cada elemento en itemList
for item in data['itemList']:
    # Verificar si 'stats' está presente en el elemento actual
    if 'stats' in item:
        # Obtener la sección 'stats' del elemento
        stats = item['stats']
        
        # Extraer los valores asociados a las claves dentro de 'stats'
        collect_count = stats.get('collectCount')
        comment_count = stats.get('commentCount')
        digg_count = stats.get('diggCount')
        play_count = stats.get('playCount')
        
        # Estructurar los datos extraídos como una tupla y agregarlos a la lista
        structured_data.append((digg_count, collect_count, comment_count, play_count))

# Ordenar la lista de datos estructurados por el valor de 'diggCount' de mayor a menor
structured_data.sort(reverse=True)

# Imprimir la lista ordenada
for data in structured_data:
    print("DiggCount:", data[0])
    print("CollectCount:", data[1])
    print("CommentCount:", data[2])
    print("PlayCount:", data[3])
    print("-----------------------------")
```
